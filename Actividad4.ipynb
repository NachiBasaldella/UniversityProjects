```python
#Actividad 4
#Solución primer enunciado
#VENTAS
def vendedor_mas_ganancias(m):
    max_ganancias = 0
    vendedor = ''
    for v in m:
        ganancias = sum(v[2:])  # Suma de las ventas para cada vendedor
        if ganancias > max_ganancias:
            max_ganancias = ganancias
            vendedor = v[0]  # Nombre del vendedor con las ganancias máximas
    return vendedor

def sucursal_menos_ganancias(m):
    sucursales = {}
    for v in m:
        sucursal = v[1]
        ganancias = sum(v[2:])  # Suma de las ventas para cada vendedor
        if sucursal in sucursales:
            sucursales[sucursal] += ganancias
        else:
            sucursales[sucursal] = ganancias
    sucursal_menos_ganancias = min(sucursales, key=sucursales.get)  # Obtiene la sucursal con las ganancias mínimas
    return sucursal_menos_ganancias

def mejor_venta(m):
    max_venta = 0
    for v in m:
        max_venta = max(max(v[2:]), max_venta)  # Encuentra la venta máxima para cada vendedor
    return max_venta

ventas = [['Mariona','Barcelona',100, 20, 30],
          ['Oscar','Sevilla',25,250,80],
          ['Roberto','Navarra',70,65,125],
          ['Julio','Barcelona',55,90,110],
          ['Sofia','Madrid',75,35,200]]

assert(vendedor_mas_ganancias(ventas) == 'Oscar')
assert(sucursal_menos_ganancias(ventas) == 'Navarra')
assert(mejor_venta(ventas) == 250)

#EVALUCIÓN DE PRIMER ENUNCIADO
#puntuación: 10
#El problema presentado tenía un nivel de dificultad adecuado para los estudiantes, lo que les permitió poner en práctica
#los conocimientos y habilidades de manera efectiva sin sentirse abrumados. La descripción del problema estaba clara y no tenía ambigüedades,
#lo que facilitó su comprensión sin la necesidad de pruebas adicionales. Los casos de prueba proporcionados eran exhaustivos y
#cubrían una amplia gama de situaciones, incluyendo los casos más complejos o límites. Estos casos de prueba demostraron ser útiles
#para validar la solución y garantizar su correcto funcionamiento.
# Además, se hizo un uso efectivo y justificado de archivos en algunos de los casos de prueba,
#lo que mejoró significativamente la calidad y la cobertura de las pruebas al simular situaciones
#de entrada y salida del mundo real.

#SOLUCIÓN SEGUNDO ENUNCIADO
import heapq
def anadir_averia(cola, descripcion, nivel_dificultad):
    # Verifica que la dificultad sea menor que 6
    if nivel_dificultad < 6:
        # Agrega la avería a la cola con prioridad, donde la prioridad es la dificultad
        heapq.heappush(cola, (nivel_dificultad, descripcion))

def anadir_lista_averias(cola, lista_averias):
    for averia in lista_averias:
        descripcion, nivel_dificultad = averia
        anadir_averia(cola, descripcion, nivel_dificultad)
    return cola

def procesar_averias(cola):
    total_averias = 0
    suma_dificultades = 0
    max_dificultad = 0

    while cola:
        dificultad, descripcion = heapq.heappop(cola)
        if dificultad >= 6:
            # Si la dificultad es igual o mayor a 6, no se atiende la avería
            continue
        total_averias += 1
        suma_dificultades += dificultad
        max_dificultad = max(max_dificultad, dificultad)

    media_dificultades = suma_dificultades / total_averias if total_averias > 0 else 0
    return total_averias, media_dificultades, max_dificultad

cola_averias = []  # Inicializa la cola de averías como una lista vacía

lista = [["Arreglar fachada", 6],
        ["Reparar fuga de agua", 3],
        ["Configurar WiFi", 2],
        ["Reemplazar bombilla", 1],
        ["Arreglar enchufe suelto", 2]]

# Añade las averías a la cola
cola_averias = anadir_lista_averias(cola_averias, lista)

# Procesa las averías y obtiene los resultados
resultado = procesar_averias(cola_averias)

# Imprime los resultados
print("Total de averías atendidas:", resultado[0])
print("Media de dificultad de las averías:", resultado[1])
print("Avería más difícil:", resultado[2])

#Evaluacion segundo enunciado
#El problema estaba dentro del nivel de dificultad esperado para los estudiantes, siendo accesible para la mayoría,
#aunque algunos podrían considerarlo demasiado complicado. La descripción del problema era en su mayoría clara,
#pero presentaba algunas áreas ambiguas. Los casos de prueba, aunque eran en su mayoría completos y abordaban diversas
#situaciones, no abordaban por completo las ambigüedades en el enunciado.


#SOLUCIÓN ENUNCIADO 3
# Función para calcular la media de tiempo de resolución de tickets
def calcular_media_tiempo_respuesta(tickets):
    return sum(ticket[2] for ticket in tickets) / len(tickets)

# Función para calcular el número total de tickets
def calcular_total_tickets(tickets):
    return len(tickets)

# Función para encontrar el ticket con el mayor tiempo de resolución y el operador que lo atendió
def encontrar_ticket_mayor_tiempo(tickets):
    return max(tickets, key=lambda x: x[2])

# Función para encontrar el ticket con el menor tiempo de resolución y el operador que lo atendió
def encontrar_ticket_menor_tiempo(tickets):
    return min(tickets, key=lambda x: x[2])

# Función para contar los tickets por operador
def contar_tickets_por_operador(tickets):
    operadores = {}
    for ticket in tickets:
        operador = ticket[1]
        if operador in operadores:
            operadores[operador] += 1
        else:
            operadores[operador] = 1
    return operadores

# Función para calcular la media del tiempo de respuesta por operador
def calcular_media_tiempo_por_operador(tickets):
    tiempos_por_operador = {}
    contador_por_operador = {}
    for ticket in tickets:
        operador = ticket[1]
        tiempo_respuesta = ticket[2]
        if operador in tiempos_por_operador:
            tiempos_por_operador[operador] += tiempo_respuesta
            contador_por_operador[operador] += 1
        else:
            tiempos_por_operador[operador] = tiempo_respuesta
            contador_por_operador[operador] = 1

    media_por_operador = {}
    for operador, tiempo in tiempos_por_operador.items():
        media_por_operador[operador] = tiempo / contador_por_operador[operador]

    return media_por_operador

# Función para contar los tickets con tiempo de respuesta mayor a 15 minutos y calcular el porcentaje
def contar_tickets_tiempo_mayor_15_min(tickets):
    total_tickets = len(tickets)
    tickets_mayor_15_min = sum(1 for ticket in tickets if ticket[2] > 15)
    porcentaje = (tickets_mayor_15_min / total_tickets) * 100
    return (tickets_mayor_15_min, porcentaje)

# Función para verificar el nivel de servicio
def nivel_servicio(tickets):
    tickets_mayor_15_min, porcentaje = contar_tickets_tiempo_mayor_15_min(tickets)
    media_tiempo_respuesta = calcular_media_tiempo_respuesta(tickets)

    if porcentaje < 10 and media_tiempo_respuesta < 15:
        return "El nivel de servicio está siendo correcto"
    else:
        return "El nivel de servicio debe mejorar y puede causar penalización"

# Ejemplo de matriz de tickets
tickets = [
    [101, "Operador1", 10],
    [102, "Operador2", 25],
    # ... (otros tickets)
]

# Salida por pantalla
print(f"Media de tiempo de resolución de tickets: {calcular_media_tiempo_respuesta(tickets):.2f} minutos")
# Imprimir otros resultados de manera similar

# Verificar el nivel de servicio
print(nivel_servicio(tickets))
Este código simplificado realiza las mismas operaciones que el código original pero de manera más concisa y legible.






#Evaluacion enunciado 3
#El problema estaba dentro del nivel de dificultad esperado para los estudiantes, siendo accesible para la mayoría,
#aunque algunos podrían considerarlo demasiado complicado. La descripción del problema era en su mayoría clara,
#pero presentaba algunas áreas ambiguas. Los casos de prueba, aunque eran en su mayoría completos y abordaban diversas
#situaciones, no abordaban por completo las ambigüedades en el enunciado.
```
